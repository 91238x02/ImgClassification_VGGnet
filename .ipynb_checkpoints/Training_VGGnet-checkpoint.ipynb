# Modeling

## VGG16

### Import module 

```python
## module import
import tensorflow as tf # tensorflow 2.0
from tensorflow.keras.models import Sequential, save_model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Conv2D, Flatten, MaxPooling2D, Activation
from tensorflow.keras.layers import Dropout, BatchNormalization, Dense
from tensorflow.keras.optimizers import Adam


from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np


## Transfer learning (전이학습)
from tensorflow.keras.applications.vgg16 import VGG16                        
from tensorflow.keras.applications.vgg19 import VGG19
# from tensorflow.keras.applications.inception_v3 import InceptionV3
# from tensorflow.keras.applications.resnet import ResNet50
```

```python
## Check tensorFlow version

print (tf.__version__)
```

### Validation data (Train:val=0.7:0.3)

```python
# 훈련/테스트 데이터를 0.7/0.3의 비율로 분리합니다.
x_train, x_val, y_train, y_val = train_test_split(X, y, 
                                                test_size = 0.3, 
                                                random_state = 777)

# Checkout
print(x_train.shape)
print(x_val.shape)  
print(y_train.shape)
print(y_val.shape)  
```

### Set Vgg16 options

```python
## VGG16 options

vgg16 = VGG16(weights = 'imagenet', include_top = False, input_shape = (224, 224, 3))
vgg16.summary()
```

```python
## 가중치 초기값 : imagenet
# layer.trainable=True : 동결 해제 (default)
# layer.trainable=False : 동결 (option)

for layer in vgg16.layers[:15]:
    layer.trainable = False
```

```python
vgg16.summary()
```

### Vgg16 뒷부분 신경망 생성

```python
# 신경망 객체 생성
model = Sequential()

# stacking vgg16
model.add(vgg16)

# Reshape : Flatten 
model.add(Flatten())

# 완전연결계층1
#model.add(Dense(256, kernel_regularizer=tf.keras.regularizers.l2(1.0)))
model.add(Dense(256, kernel_regularizer=tf.keras.regularizers.l2(1.0)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

# 완전연결계층2
model.add(Dense(256, kernel_regularizer=tf.keras.regularizers.l2(1.0)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.5))

# 출력층(softmax)
#model.add(Dense(10))  # class : 10
model.add(Dense(5))  # class : 5
model.add(Activation('softmax'))  

# Check out model 
model.summary()
```

### Compile

```python
model.compile(loss='categorical_crossentropy', 
              optimizer=Adam(lr = 0.0001), 
              metrics=['accuracy'])
```

### fitting

```python
## Parameters
epochs = 30
batch_size=50
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=1)  #Early stopping
#https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping
    

## Fitting
hist = model.fit(x_train, y_train, 
                 validation_data=(x_val, y_val), 
                 epochs=epochs, 
                 batch_size=batch_size,
                 callbacks = [callback])
```

```python
# ## Check out accuracy (with independent data)
# scores3 = model.evaluate(ind_X, ind_y, verbose=0)

# print("Vgg16 ind_dataset Error : %.2f%%" % (100-scores3[1]*100))
# print("Vgg16 ind_dataset loss : %.2f" % (scores3[0]))
```

### Visualization

### Results

```python
## Check out accuracy

scores = model.evaluate(x_train, y_train, verbose=0)
scores2 = model.evaluate(x_val, y_val, verbose=0)

print("Vgg16 train Error : %.2f%%" % (100-scores[1]*100))
print("Vgg16 val Error : %.2f%%" % (100-scores2[1]*100))
print(' ')
print("Vgg16 train Loss : %.2f" % (scores[0]))
print("Vgg16 val Loss : %.2f" % (scores2[0]))
```

# Extraction of Convolution Layer 

```python
"""
합성곱 층의 filter 가 적용된 결과가 어떠한지를 사람의 눈으로 확인할 수 있도록 확인하는 코드
현재까지의 구현으로는 출력이 가능하긴 한데 심층부로 갈수록 너무 많이 나와서 일자 형태의 bar 로 나타나서 확인이 불가능.
따라서 모든 filter 결과를 보는 것이 아니라 적은 수의 값들만 보는 함수를 사용하는 것이 좋아보임.
또한 위 합성곱 extraction 을 보고 모델 커스텀의 방향을 결정하는 것은 인간의 주관이 들어가는 결과를 초래할 수 있기 때문에
위 Layer 을 직접 확인하기 보다는 합성곱 출력값들 중 이상치를 확인하는 함수나 방법을 사용하는 것이 더 추천됨.
"""
```

```python
model.summary()
```

```python
model.layers[0].summary()
```

```python
k = model.layers[0]
layer_outputs = [layer.output for layer in k.layers]
layer_outputs
```

```python
layer_names = [layer.name for layer in k.layers]
layer_names
```

```python
feature_map_model = tf.keras.models.Model(k.input,layer_outputs)
feature_map_model.summary()
```

```python
x = X[0]
x=x.reshape(1,224,224,3)
x.shape
```

```python
feature_maps = feature_map_model.predict(x)
```

```python
for layer_name, feature_map in zip(layer_names, feature_maps):
    print(f"The shape of the {layer_name} is =======>> {feature_map.shape}")
```

```python
for layer_name, feature_map in zip(layer_names, feature_maps):  
    if len(feature_map.shape) == 4:
        k = feature_map.shape[-1]  
        size=feature_map.shape[1]
        image_belt = np.zeros((size, size*k))
        
        for i in range(k):
            feature_image = feature_map[0, :, :, i]
            feature_image-= feature_image.mean()
            feature_image/= feature_image.std ()
            feature_image*= 64
            feature_image+= 128
            feature_image= np.clip(feature_image, 0, 255).astype('uint8')
            image_belt[:, i * size : (i + 1) * size] = feature_image 
            
        scale = 20. / k
        plt.figure( figsize=(scale * k, scale) )
        plt.title ( layer_name )
        plt.grid  ( False )
        plt.imshow( image_belt, aspect='auto')
```
